```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
%matplotlib inline
```

```python
qr = QuantumRegister(1)
cr = ClassicalRegister(1)
circuit = QuantumCircuit(qr, cr)
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend=simulator).result()
statevector = result.get_statevector()
print(statevector)
array_to_latex(statevector)
```

```python
plot_bloch_multivector(statevector)
```

```python
circuit.h(qr)
result = execute(circuit, backend=simulator).result()
statevector_with_hadmard_gate = result.get_statevector()
print(statevector_with_hadmard_gate)
array_to_latex(statevector_with_hadmard_gate)
```

```python
plot_bloch_multivector(statevector_with_hadmard_gate)
```

```python
# measure the circuit with Classical register you will get sometime |1> or |0>
circuit.measure(qr, cr)
result = execute(circuit, backend=simulator).result()
statevector_with_hadmard_gate_measurement = result.get_statevector()
print(statevector_with_hadmard_gate_measurement)
array_to_latex(statevector_with_hadmard_gate_measurement)
```

```python
plot_bloch_multivector(statevector_with_hadmard_gate_measurement)
```
